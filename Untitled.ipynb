```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from pylab import *
```

```python
def main():
    path = ("C:/Users/welcome/Desktop/tree1.jfif")
    
	# read the image from the specified path
    img = cv2.imread(path,0)
    
	# applying histogram equalization on the given input image
    output1 = cv2.equalizeHist(img)
    
	
	#displaying the image with enhanced outputs
    output = [img,output1]
    titles = ['Original Image', 'Enhanced image']
    
    for i in range(2):
        plt.subplot(1,3,i+1)
		
		#converting the images into gray scale
        plt.imshow(output[i],cmap = 'gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
    
	# calculating histogram for the original image
    histr = cv2.calcHist([img],[0],None,[256],[0,256])
	
	# calculating histogram for the output image
    outhistr1 = cv2.calcHist([output1],[0],None,[256],[0,256])
	
	# plotting histogram using matplotlib
    plt.plot(histr)
    plt.plot(outhistr1)
    plt.show()
    
	
	
    
```

```python
if __name__ == "__main__":
    main()
```

```python
def main():
#morphological opening filter
    black = (30,30,30)
    white = (255,255,255)
    threshold = (100,100,100)

# Open input image in grayscale mode and get its pixels.
    img = Image.open("C:/Users/welcome/Desktop/tree1.jfif").convert("LA")
    pixels = img.getdata()

    newPixels = []

# Compare each pixel 
    for pixel in pixels:
        if pixel < threshold:
            newPixels.append(black)
        else:
            newPixels.append(white)

# Create and save new image.
    newImg = Image.new("RGB",img.size)
    newImg.putdata(newPixels)
    newImg.save("newImage.jpg")
```

```python
if __name__ == "__main__":
    main()
```

```python
def main():
    
    path = "C:/Users/welcome/Desktop/tree1.jfif"
    imgpath =  path 

    img = cv2.imread(imgpath, 1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    R, G, B = cv2.split(img)
    
    output1_R = cv2.equalizeHist(R)
    output1_G = cv2.equalizeHist(G)
    output1_B = cv2.equalizeHist(B)
    
    output1 = cv2.merge((output1_R, output1_G, output1_B))

#    clahe = cv2.createCLAHE()
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

    output2_R = clahe.apply(R)
    output2_G = clahe.apply(G)
    output2_B = clahe.apply(B)
    
    output2 = cv2.merge((output2_R, output2_G, output2_B))


    output = [img, output1, output2]
    titles = ['Original Image', 'Adjusted Histogram', 'CLAHE']
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(output[i])
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])
    plt.show()
```

```python
if __name__ == "__main__":
    main()
```
